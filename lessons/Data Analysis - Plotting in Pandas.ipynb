# Cheat Sheet

**Simple plotting**  
`df.plot()`  

**Advanced plotting**  
`fig, axes = plt.subplots(nrows=n, ncols=n, figsize=(n, n))`  

**Plot type**  
`df.plot(kind='line', 'bar', 'barh', 'kde', 'density')`  

**Set title**  
`set_title('type title here')`  

**Set y label**  
`set_ylabel('type y axis label')`  

**Set x label**  
`set_xlabel('type x axis label')`  

**Show legend**  
`legend()`  

**Set legend labels**  
`df.plot(label='label name')`  
`axes.legend(["label name"], loc='best')`  

```python
# Sample dataframe
d = {'date': [pd.to_datetime('1/1/2013'), pd.to_datetime('1/1/2014'), pd.to_datetime('1/1/2015')],
     'number': [1,2,3],
     'letter': ['A','B','C']}
df = pd.DataFrame(d)
df
```

```python
# simple plotting
df.plot();
```

```python
# pretty plotting
p = df['number'].plot(label='label name')
p.set_title('This is the title')
p.set_ylabel('the y axis')
p.set_xlabel('the x axis')
p.legend(loc='best');
```

```python
# Two simple plots
df.plot()
df.plot();
```

```python
# Advanced Plotting
####################
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(13, 10))
fig.subplots_adjust(hspace=2.0) ## Create space between plots

# Chart 1
df.plot(ax=axes)

# add a little sugar
axes.set_title('This is the title')
axes.set_ylabel('the y axis')
axes.set_xlabel('the x axis')
axes.legend(["label goes here"], loc='best');
```

```python
# Advanced Plotting
####################
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(13, 10))
fig.subplots_adjust(hspace=2.0) ## Create space between plots

# Chart 1
df.plot(ax=axes[0])

# Chart 2
df.set_index('date')['number'].plot(ax=axes[1])

# add a little sugar
axes[0].set_title('This is the title')
axes[0].set_ylabel('the y axis')
axes[0].set_xlabel('the x axis')
axes[0].legend(["label chart 1"], loc='best')
axes[1].legend(["label chart 2"], loc='best');
```

```python
# Advanced Plotting
####################
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13, 10))
fig.subplots_adjust(hspace=0.25) ## Create space between plots

# Chart 1
df.plot(ax=axes[0,0])

# Chart 2
df.set_index('date')['number'].plot(ax=axes[0,1])

# Chart 3
df.plot(ax=axes[1,0], color='r')

# Chart 4
df.plot(ax=axes[1,1], color='g', style='--')

# add a little sugar
axes[0,1].set_title('This is the title')
axes[0,1].set_ylabel('the y axis')
axes[0,1].set_xlabel('the x axis')
axes[0,0].legend(["label chart 1"], loc='best')
axes[0,1].legend(["label chart 2"], loc='best')
axes[1,0].legend(["label chart 3"], loc='best')
axes[1,1].legend(["label chart 4"], loc='best');
```
